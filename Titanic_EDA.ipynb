### Data Cleaning

```python
#Check the null values
df.isnull().sum()
```

```python
# 1. handle missing age values
# will use median to fill the missing age values
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

df['Age'].isnull().sum()
```

```python
# 2. Filling Embarked with mode
mode_Embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_Embarked)

df['Embarked'].isnull().sum()
```

```python
# 3. handle the cabin column
# 77% of cabin column have missing data , will create a new column as 'Has_cabin'
df['Has_cabin'] = df['Cabin'].notna().astype(int)


df.drop('Cabin', axis=1, inplace=True) #This will drop the original the original cabin column
```

### Univariate analysis

```python
print("Analysing categorical feature")

fig, axes = plt.subplots(2,3, figsize=(18,12) )
fig.suptitle('Unvariate analysis of categorical feature', fontsize= 16)

sns.countplot(ax=axes[0,0], x= 'Survived',data=df).set_title('Survival distribution')
sns.countplot(ax=axes[0,1], x='Pclass', data=df).set_title('Passenger class distribution')
sns.countplot(ax=axes[0,2], x='Sex',data= df).set_title('Gender distribution')
sns.countplot(ax=axes[1,0], x='SibSp', data=df).set_title('Siblings/Spouses Aboard')
sns.countplot(ax=axes[1,1], x='Parch', data=df).set_title('Parents/Children Aboard')
sns.countplot(ax=axes[1,2], x='Embarked',data= df).set_title('Port of Embarkation')

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

###Feature Engineering

```python
df['FamilySize'] =df['SibSp'] + df['Parch'] + 1

df['IsAlone'] = 0
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1

df[['FamilySize','IsAlone']].head()
```

```python
fig, axes =plt.subplots(1,2,figsize=(16,6))

sns.barplot(ax=axes[0],x='FamilySize',y='Survived', data= df).set_title('Survival rate by family size')
sns.barplot(ax=axes[1], x='IsAlone', y='Survived',data= df).set_title('Survival rate by those travelling alone')
plt.show()
```

```python
# Extract title
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.',expand=False)
df['Title'].value_counts()
```

```python
df['Title'] = df['Title'].replace(['Dr','Rev','Col','Don','Lady','Sir','Capt','Countess','Jonkheer','Major'],'Rare')

df['Title'] = df['Title'].replace(['Mlle','Ms'],'Miss')
df['Title'] = df['Title'].replace('Mme','Mrs')

plt.figure(figsize=(12,6))
sns.barplot(x='Title',y='Survived',data=df)
plt.title('Survival Rate by Title')
plt.ylabel('Survival Probability')
plt.show()
```

###Multivariate Analysis

```python
sns.catplot(x='Pclass',y='Survived', hue='Sex',kind='bar',data=df,height=6,aspect=1.5)
plt.title('Survival rate by sex and Pclass')
plt.ylabel('Survival probablity')
plt.show()
```

###Correlation analysis


```python
# Heatmap
plt.figure(figsize=(14,10))
numeric_col =df.select_dtypes(include=np.number)
correlation_matrix = numeric_col.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f')
plt.title('Correlation matrix of numerical feature')
plt.show()
```

# Overall Finding
1. Women survived much more than men.

2. First-class passengers had the highest survival, third-class the lowest.

3. Children had better survival chances than adults.

4. Passengers who paid higher fares survived more often.

5. People from Cherbourg (C) survived more because many were first-class.

6. Having a cabin number (higher status) increased survival chances.
